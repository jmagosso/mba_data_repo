```python
nota1 = nota2 = 0.0
media = 0.0

nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
media = (nota1 + nota2) / 2

print("A média é: ", media)

if media >= 7:
    print("Aprovado")
elif(media >= 5): 
    print("Recuperação")
else:
    print("Reprovado")

print("Sua média é: {}".format(media))
```
