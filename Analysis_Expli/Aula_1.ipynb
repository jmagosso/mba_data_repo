#APLICAR A MELHOR MANEIRA DE EU ENTENDER

#fazer as minhas análises estátisticas


```python
import numpy as np
from scipy.stats import mode


#Verify the data structure
print("Colunas disponíveis:", data.columns)
print(data.head())

# # Select the spending column
valores = data["gastos_online (USD)"]

# # 1. Arithmetic Mean
media_aritmetica = valores.sum() / len(valores)


# # 2. Geometric Mean
valores_positivos = valores[valores > 0]
media_geometrica = np.exp(np.log(valores_positivos).sum() / len(valores_positivos)) if len(valores_positivos) > 0 else np.nan

# # 3. Weighted Mean (using equal weights as placeholder)
pesos = np.ones(len(valores))
media_ponderada = (valores * pesos).sum() / pesos.sum()

# # 4. Harmonic Mean
media_harmonica = len(valores_positivos) / (1 / valores_positivos).sum() if len(valores_positivos) > 0 else np.nan

# # 5. Median
valores_ordenados = np.sort(valores)
print(valores_ordenados)
n = len(valores_ordenados)
mediana = (valores_ordenados[n//2 - 1] + valores_ordenados[n//2]) / 2 if n % 2 == 0 else valores_ordenados[n//2]

# # 6. Mode
moda_valor = mode(valores, keepdims=False)[0]

# # 7. Midpoint
ponto_medio = (valores.min() + valores.max()) / 2

# # 8. Mean of Frequency Distribution (Fixed)
freq_table = valores.value_counts().sort_index()
valores_unicos = freq_table.index.to_numpy()  # Convert Index to NumPy array
frequencias = freq_table.values
media_freq = (valores_unicos * frequencias).sum() / frequencias.sum()

# # 9. Quartiles
q1 = np.percentile(valores, 25)
q2 = np.percentile(valores, 50)
q3 = np.percentile(valores, 75)


resultados = {
        "Média Aritmética": media_aritmetica,
         "Média Geométrica": media_geometrica,
         "Média Ponderada": media_ponderada,
         "Média Harmônica": media_harmonica,
         "Mediana": mediana,
         "Moda": moda_valor,
         "Ponto Médio": ponto_medio,
         "Média de Frequências": media_freq,
         "Quartis (Q1, Q2, Q3)": [q1, q2, q3]
    }





```

```python
grupos = data.groupby("Idade")

resultados = {}

for nome_grupo, grupo in grupos:
    valores = grupo["gastos_online (USD)"]
    if len(valores) == 0:
      continue


    # # 1. Arithmetic Mean
    media_aritmetica = valores.sum() / len(valores)

    # # 2. Geometric Mean
    valores_positivos = valores[valores > 0]
    media_geometrica = np.exp(np.log(valores_positivos).sum() / len(valores_positivos)) if len(valores_positivos) > 0 else np.nan

    # # 3. Weighted Mean (using equal weights as placeholder)
    pesos = np.ones(len(valores))
    media_ponderada = (valores * pesos).sum() / pesos.sum()

    # # 4. Harmonic Mean
    media_harmonica = len(valores_positivos) / (1 / valores_positivos).sum() if len(valores_positivos) > 0 else np.nan

    # # 5. Median
    valores_ordenados = np.sort(valores)
    n = len(valores_ordenados)
    mediana = (valores_ordenados[n//2 - 1] + valores_ordenados[n//2]) / 2 if n % 2 == 0 else valores_ordenados[n//2]

    # # 6. Mode
    moda_valor = mode(valores, keepdims=False)[0]

    # # 7. Midpoint
    ponto_medio = (valores.min() + valores.max()) / 2

    # # 8. Mean of Frequency Distribution (Fixed)
    freq_table = valores.value_counts().sort_index()
    valores_unicos = freq_table.index.to_numpy()  # Convert Index to NumPy array
    frequencias = freq_table.values
    media_freq = (valores_unicos * frequencias).sum() / frequencias.sum()

    # # 9. Quartiles
    q1 = np.percentile(valores, 25)
    q2 = np.percentile(valores, 50)
    q3 = np.percentile(valores, 75)


    resultados[nome_grupo] = {
        "Média Aritmética": media_aritmetica,
        "Média Geométrica": media_geometrica,
        "Média Ponderada": media_ponderada,
        "Média Harmônica": media_harmonica,
        "Mediana": mediana,
        "Moda": moda_valor,
        "Ponto Médio": ponto_medio,
        "Média de Frequências": media_freq,
        "Quartis (Q1, Q2, Q3)": [q1, q2, q3]
    }


# Display results by age group
for grupo, stats in resultados.items():
    print(f"\nFaixa Etária: {grupo}")
    for medida, valor in stats.items():
        if medida == "Quartis (Q1, Q2, Q3)":
            print(f"{medida}: [{valor[0]:.2f}, {valor[1]:.2f}, {valor[2]:.2f}]")
        else:
            print(f"{medida}: {valor:.2f}")

```
