```python
# Cria a tabela 'clientes' no schema 'bronze' do catálogo 'data_catalog_01_d' com estrutura definida, usando Delta Lake

spark.sql("""
CREATE TABLE IF NOT EXISTS data_catalog_01_d.bronze.clientes (
    id BIGINT ,
    cnpj STRING,
    razao_social STRING,
    fantasia STRING,
    cidade STRING,
    endereco STRING,
    numero STRING,
    estado STRING,
    vendedor STRING,
    data_carga TIMESTAMP
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS data_catalog_01_d.silver.customers (
    customer_id STRING COMMENT 'Identificador único do cliente',
    document_number STRING COMMENT 'Número do documento formatado (CNPJ)',
    company_name STRING COMMENT 'Razão social da empresa',
    nick_name STRING COMMENT 'Nome fantasia da empresa',
    sales_person STRING COMMENT 'Nome do vendedor responsável',
    address STRING COMMENT 'Endereço do cliente',
    address_number STRING COMMENT 'Número do endereço',
    city_name STRING COMMENT 'Nome da cidade',
    state STRING COMMENT 'Estado',
    insert_date TIMESTAMP COMMENT 'Data de inserção do registro',
    update_date TIMESTAMP COMMENT 'Data de última atualização do registro'
)
USING DELTA
""")
```
